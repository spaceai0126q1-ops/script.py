```python
import os
import pandas as pd

# Input and output file names
input_file = "s1_DAILY_MOVEMENT_LOG_01MAY2026_test1.xlsx"
output_folder = "DAILY_MOVEMENT_LOG_MAY2026"
output_file = os.path.join(output_folder, "DAILY_MOVEMENT_LOG_01MAY2026_test1_modified.xlsx")
log_file = os.path.join(output_folder, "missing_columns_log.txt")

# Ensure the output folder exists
os.makedirs(output_folder, exist_ok=True)

# Load the Excel file
df = pd.read_excel(input_file)

# Columns to extract
columns_to_extract = ["ACID", "TYPE", "EQUIP", "DEP_APT", "ARR_APT", "ACFT_REG", "ENTRY_DATE", "EXIT_DATE", "FIR_ROUTE"]

# Check which columns exist
available_columns = [col for col in columns_to_extract if col in df.columns]
missing_columns = [col for col in columns_to_extract if col not in df.columns]

# Inform user about missing columns
if missing_columns:
    print(f"Warning: The following columns are missing and will not be included: {missing_columns}")
    # Write missing columns to log file
    with open(log_file, "w") as f:
        f.write("Missing columns:\n")
        for col in missing_columns:
            f.write(f"- {col}\n")
    print(f"Missing columns logged in: {log_file}")

# Filter the dataframe with available columns only
df_filtered = df[available_columns]

# Save the filtered dataframe to a new Excel file
df_filtered.to_excel(output_file, index=False)

print(f"Modified file saved at: {output_file}")
```
