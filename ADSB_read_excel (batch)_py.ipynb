## a. ADSB_READ_EXCEL (batch)

```python
import os
import pandas as pd

# Folder and output setup
input_folder = "DAILY_MOVEMENT_LOG_16-31MAY2026"
output_file = os.path.join(input_folder, "DAILY_MOVEMENT_LOG_MAY2026_compiled.xlsx")
log_file = os.path.join(input_folder, "missing_columns_log.txt")

# Ensure folder exists
os.makedirs(input_folder, exist_ok=True)

# Columns to extract
columns_to_extract = ["ACID", "TYPE", "EQUIP", "DEP_APT", "ARR_APT", "ACFT_REG", "ENTRY_DATE", "EXIT_DATE", "FIR_ROUTE"]

# Dates to process (16-31 May 2026)
dates = range(16, 32)

# Open log file
with open(log_file, "w") as log:
    log.write("Missing columns log (16-31 May 2026):\n\n")

    # Create ExcelWriter for compiled output
    with pd.ExcelWriter(output_file, engine="openpyxl") as writer:
        for day in dates:
            # Construct filename
            input_file = f"DAILY_MOVEMENT_LOG_{day:02d}MAY2026.xlsx"
            input_path = os.path.join(input_folder, input_file)

            try:
                # Load file
                df = pd.read_excel(input_path)

                # Check available/missing columns
                available_columns = [col for col in columns_to_extract if col in df.columns]
                missing_columns = [col for col in columns_to_extract if col not in df.columns]

                # Log missing columns
                if missing_columns:
                    log.write(f"{input_file} - Missing: {missing_columns}\n")

                # Filter and save to sheet
                df_filtered = df[available_columns]
                sheet_name = f"May{day}"
                df_filtered.to_excel(writer, sheet_name=sheet_name, index=False)

                print(f"Processed {input_file} ? Sheet: {sheet_name}")

            except FileNotFoundError:
                log.write(f"{input_file} - File not found\n")
                print(f"Warning: {input_file} not found, skipped.")

print(f"\nCompiled file saved at: {output_file}")
print(f"Missing columns log saved at: {log_file}")
```
